## Basic usage example

```python
import MDAnalysis as mda
import numpy as np
from MDAnalysis.tests.datafiles import PRM_NCBOX, TRJ_NCBOX

from transport_analysis.viscosity import ViscosityHelfand as VH
```

```python
# # test data with all required info (masses, velocities, positions, box volume)
u = mda.Universe(PRM_NCBOX, TRJ_NCBOX)
ag = u.select_atoms("name O and resname WAT and resid 1-10")
```

```python
t_visc = VH(ag).run()
print(t_visc.results.timeseries)
```
